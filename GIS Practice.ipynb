```python
import matplotlib.pyplot as plt
```

```python
import geopandas as gpd
```

```python
cities = gpd.read_file(r"Shapefiles\belgian_cities.shp")
```

```python
cities.plot(cmap = 'jet', column='NAME_4', figsize = (10, 10))
```

```python
AOI = gpd.read_file(r"Shapefiles\area_of_interest_.shp")
```

```python
AOI.plot()
```

```python
fig, ax = plt.subplots(1)
cities.plot(ax=ax, cmap='jet')
AOI.plot(ax=ax, facecolor='yellow')
```

```python
cities_in_AOI = gpd.overlay(cities, AOI, how='intersection')
cities_in_AOI.plot(figsize=(6,6), cmap='jet', column='NAME_4')
```
